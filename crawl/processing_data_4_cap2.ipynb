```python
import pandas as pd
data = pd.read_csv(r'D:\virtual environment\Thesis\Thesis\crawl\data\finance_data\balance_sheet_20251008_182301_3.csv')

# Giả sử data đã có sẵn
N = len(data)
target = 4_000_000

# Nhân bản cho đủ 4 triệu dòng
repeat_times = int(target / N) + 1
df_big = pd.concat([data] * repeat_times, ignore_index=True)

# Gán cột years là unique (ví dụ 1 đến 4_000_000)
df_big['years'] = range(1, target + 1)

print(df_big.shape)
display(df_big.head(20))  # Kiểm tra: True
```
